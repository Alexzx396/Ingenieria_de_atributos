```python
#CARGAR LIBRERIAS
import pandas as pd
import numpy as np
import datetime as dt
import calendar
import plotly.express as px

#librerias visuales
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

### Hago una breve inspección del dataset

```python
#CARGAR DATASET
df_Carsales=pd.read_excel('Carsales_b.xlsx')
df_Carsales.head()
```

```python
df_Carsales.info()
```

```python
df_Carsales.dropna(inplace=True)
```

```python
df_Carsales.nunique()
```

### Luego genero una copia del dataframe original, dónde voy a codificar las variables tipo object para que sean categoricas y luego asignarles valores numéricos

```python
df=df_Carsales.copy()
```

```python
obj=list((df.dtypes=='object').index)
df[obj]=df[obj].astype('category')
df.head()
```

```python
for col in df[obj].columns:
    df[f'{col}']=df[f'{col}'].cat.codes
```

```python
df.head()
```

### Ahora que el dataframe tiene sólo int o float, podemos entrenar el random forest con todas las variables para encontrar las más relevantes

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
X=df.drop('Price',axis=1)
y=df.Price
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=123)
```

```python
rfr=RandomForestRegressor(max_depth=4,random_state=123)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred=rfr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

### Como ven, la métrica es buena y son pocas variables que realmente influyen. Les sugiero usar sólo aquellas que no tienen puros ceros para entrenar de aquí en adelante

```python
pd.DataFrame.from_dict({'Variable':X.columns,'Importancia':rfr.feature_importances_})
```
