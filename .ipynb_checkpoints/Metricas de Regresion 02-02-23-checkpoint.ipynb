```python
#Importacion ded librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import max_error
y_verdadero = [2, 5, 3, 1, 6]
y_predicho = [2, 5, 3, 1, 0]

max_error(y_verdadero, y_predicho) #distancia maxima absoluta 

#donde usar? ejemplo limite de credito que se la da a un cliente
```

```python
#MAE Median Absolute error
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_verdadero, y_predicho)

#igual que la de arriba pero promediando entre todos los valores
```

```python
#MSE Mean Square Error - Error cuadratico medio
from sklearn.metrics import mean_squared_error

#da mas peso a los errores grandes que a los chicos por el cuadrado.

mean_squared_error(y_verdadero, y_predicho)
```

```python
#RSS Suma de los cuadradros de los residudos. (No la tiene sklearn)

mean_squared_error(y_verdadero, y_predicho)*len(y_predicho)
```

```python
#RMSE Raiz cuadrada del error cuadratico medio. Incorpora la raiz para que el error este en la misma unidad de medida

#MSE Mean Square Error - Error cuadratico medio
from sklearn.metrics import mean_squared_error

mean_squared_error(y_verdadero, y_predicho, squared=False)
```

```python
# R2 Coeficiente de determinacion - evalua que tan bien se ajusta nuestro modelo a nuestros datos
#OJO ES SOLO PARA LINEAL!

from sklearn.metrics import r2_score

y_verdadero = [2, 5, 3, 1, 6]
y_predicho = [2, 5, 3, 1, 6]

r2_score(y_verdadero, y_predicho)
```

```python
#Importacion de los datos
dataset = pd.read_csv("student_scores.csv", sep = ",")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
#Ploteamos el dataset
dataset.plot(x='Hours', y='Scores', style="o")
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
#Preparacion de datos
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
#Empezamos a crear nuestro modelo
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Entrenando el modelo
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Recuperamos la intersección
print(regressor.intercept_)
```

```python
#La pendiente
print(regressor.coef_)
```

```python
#Hacemos nuestras predicciones
y_pred = regressor.predict(X_test)
y_pred
```

```python
#Convertimos en df la salida
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

Evaluación del modelo:
El último paso es evaluar el rendimiento del algoritmo. Este paso es particularmente importante para comparar qué tan bien funcionan los diferentes algoritmos en un conjunto de datos en particular. Para los algoritmos de regresión, se utilizan comúnmente tres métricas de evaluación:
El error absoluto medio (MAE)
El error cuadrático medio (MSE)
Root Mean Squared Error (RMSE)

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
#Carga y muestra del dataset
automobile_df = pd.read_csv("auto-mpg.csv", sep = ",")

# Ver 5 registros aleatorios
automobile_df.sample(5)
```

```python
#Tamaño del dataset
automobile_df.shape 
```

```python
#Info
automobile_df.info()
```

```python
#Preparacion de datos
automobile_df = automobile_df.replace('?', np.nan)
automobile_df = automobile_df.dropna()
automobile_df.shape
```

```python
#Eliminamos columnas inncesarias
automobile_df.drop(['origin', 'car name'], axis=1, inplace=True)
```

```python
#Vemos registros aleatorios
automobile_df.sample(5)
```

```python
#Info
automobile_df.info()
```

```python
#Convertimos el formato de ‘model year’ a año completo
automobile_df['model year'] = '19' + automobile_df['model year'].astype(str)
```

```python
#Vemos los datos
automobile_df.sample(5)
```

```python
#Agregamos la columna age
automobile_df['age'] = datetime.datetime.now().year - pd.to_numeric(automobile_df['model year'])
```

```python
#Eliminamos model year
automobile_df.drop(['model year'], axis=1, inplace=True)
```

```python
#Corroboramos que todo este bien
automobile_df.sample(5)
```

```python
#Vemos los tipos de datos
automobile_df.dtypes
```

```python
#Convertimos horsepower en numerico
automobile_df['horsepower'] = pd.to_numeric(automobile_df['horsepower'], errors='coerce')
```

```python
#Analisis estadistico de datos
automobile_df.describe()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(automobile_df['displacement'], automobile_df['mpg'])

plt.xlabel('Desplazamiento')
plt.ylabel('Millas por galon')
```

```python
#Regresion Lineal
from sklearn.model_selection import train_test_split

X = automobile_df[['horsepower']]
Y = automobile_df['mpg']

#Train y Test
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#Importamos, creamos y fiteamos el modelo
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
#Realizamos las predicciones
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

print('Puntaje Testing: ', r2_score(y_test, y_pred))
```

```python
#Ploteamos 
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')

plt.xlabel('Caballos de Fuerza')
plt.ylabel('Mpg')
plt.show()
```

```python
#Regresion lineal con age
X = automobile_df[['age']]
Y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

linear_model = LinearRegression(normalize=True).fit(x_train, y_train)

print('Puntaje de entrenamiento: ', linear_model.score(x_train, y_train))

y_pred = linear_model.predict(x_test)

print('Puntaje de Testing: ', r2_score(y_test, y_pred))
```

```python
#Ploteamos 
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')

plt.xlabel('Age')
plt.ylabel('Mpg')
plt.show()
```

```python
#Regresion lineal con varias caracteristicas
X = automobile_df[['displacement', 'horsepower', 'weight']]
Y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#Creamos el modelo
linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
print('Training score: ', linear_model.score(x_train, y_train))
```

```python
#Coeficientes
predictors = x_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()

print(coef)
```

```python
#Realizamos las predicciones
y_pred = linear_model.predict(x_test)
```

```python
#Ploteamos
plt.figure(figsize = (20,10))

plt.plot(y_pred, label='Prediccion')
plt.plot(y_test.values, label='Real')

plt.ylabel('Mpg')

plt.legend()
plt.show()
```

```python
#Evaluación de Modelos
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Ejemplo 5

```python
import pandas as pd                         
import statsmodels.api as sm               
import statsmodels.formula.api as smf      
import matplotlib.pyplot as plt 

#Link de Interés: https://www.statsmodels.org/devel/
```

```python
#Carga de datos
file = 'https://raw.githubusercontent.com/fhernanb/Python-para-estadistica/master/03%20Regression/Regresi%C3%B3n%20lineal%20simple/softdrink.csv'
df = pd.read_csv(file)
```

```python
#Vemos nuestros datos
df.head()  
```

```python
#Scatterplot
df.plot(kind='scatter', x='x1', y='y')
```

```python
#Creacion del modelo con statsmodels.api.

X = df["x1"] # X: variables independientes
X = sm.add_constant(X)# Se agrega una intercepción (beta_0) al modelo
y = df["y"] # Y: variable respuesta (o variable dependiente)

# Fiteamos el modelo
mod1 = sm.OLS(y, X).fit()   
predictions = mod1.predict(X)
```

```python
#Estadisticas del modelo
mod1.summary()   
```

```python
#Ajuste del modelo - statsmodels.formula.api
mod2 = smf.ols('y ~ x1', data=df).fit()
```

```python
#Vemos las estadisticas del modelo
print(mod2.summary())
```

```python
#Recta de regresion estimada
fig, ax = plt.subplots(figsize=(8,6))                                    # Ajuste del gráfico
ax.plot(df["x1"] , df["y"] , 'o', label="Datos")                         # Ajuste para las etiquetas del gráfico
ax.plot(df["x1"], mod2.fittedvalues, 'r--.', label="Ajustado")           # Ajuste para las etiquetas del gráfico
legend = ax.legend(loc="best")                                           # Ajustes para las leyendas del gráfico
plt.xlabel('Cantidad de cajas')                                          # Etiqueta del eje X
plt.ylabel('Tiempo (min)')                                               # Etiqueta del eje Y
plt.title('Diagrama de dispersión con la recta del modelo ajustado');    # Título del gráfico
```
