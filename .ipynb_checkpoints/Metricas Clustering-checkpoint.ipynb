```python
# Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Mall_Customers.csv", sep = ",")
dataset.head()
```

```python
#Nos quedamos con las variables 3 y 4 de conjunto de datos, las cuales corresponden al ingreso anual en miles y la puntuaci√≥n del cliente.
X = dataset.iloc[:, [3, 4]].values
```

```python
# Metodo del Codo para encontrar el numero optimo de clusters

from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Grafica de la suma de las distancias
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

#Link de Inter√©s: https://es.wikipedia.org/wiki/K-medias
```

En el bloque anterior, generamos los clusters para valores de 1 a 10 (en el rango de 1 a 11) y obtenemos para cada uno de ellos, la suma de las distancias con el tributo inertia_ del objeto kmeans. En la gr√°fica observamos que la disminuci√≥n en la suma de las distancias se aten√∫a cuando el n√∫mero de clusters es igual a 5, por lo que, para este caso pr√°ctico, el n√∫mero √≥ptimo de clusters ser√° de 5.

```python
# Creando el k-Means para los 5 grupos encontrados
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42) #k-means++ acelera la convergencia, respecto de random
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualizacion grafica de los clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

# Silhouette

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X, kmeans.labels_, metric='euclidean')
print('Silhouetter Score: %.3f' % score)

#mide la calidad del agrupamiento a mayor valor mejor. 
```

* Se calcula (B-A) / MAX(A,B) donde a es el promedio de las distancias entre una muestra y todolos demas puntos que pertencen a la misma clase
Y B la distancia entre la muestra y todos los puntos del siguiete lcuster mas cercano.
* Toma valores entre -1 y 1 
* Interpretacion: 
    Valores cercanos a -1 indican un agrupamiento incorrecto.
    Valores cercanos a 0 indican clusters traslapados. (no logramos separar bien)
    Valores cercanos a 1 indican clusters altamente densos.

```python
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8)) #Definimos el tama√±o del gr√°fico
for i in [2, 3, 4, 5]:
    
    #Creando diferentes instancias de kmeans para distinta cantidad de cluster
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)

    #Visualizamos el √≠ndice Silhouette con yellowbrick
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X)
```

Con la aplicaci√≥n del m√©todo de Silhouette, podemos corroborar que el n√∫mero K = 5 es el valor m√°s √≥ptimo para nuestra clusterizaci√≥n, hecho que confirma la aplicaci√≥n del m√©todo del codo (Elbow).

# Ejemplo 2

```python
#Importamos las libreer√≠as
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib inline
```

```python
#Generamos data aleatoria
X= np.random.rand(50,2)
Y= 2 + np.random.rand(50,2)
Z= np.concatenate((X,Y))
Z=pd.DataFrame(Z)
```

```python
#Ploteamos
sns.scatterplot(Z[0],Z[1])
```

```python
#Aplicamos K-Means con 2 cl√∫sters
KMean= KMeans(n_clusters=2)
KMean.fit(Z)
label=KMean.predict(Z)
```

```python
#Calculamos el √çndice de Silhouette
print(f'Silhouette Score(n=2): {silhouette_score(Z, label)}')
```

Insight
Podemos decir que los grupos est√°n bien separados entre s√≠ ya que la m√©trica de Silhouette nos arroja un resultado cercano a 1. Sin embargo, para comprobar si nuestra puntuaci√≥n de silueta proporciona la informaci√≥n correcta o no, podemos crear otro diagrama de dispersi√≥n que muestre los puntos de datos etiquetados como se visualiza a continuaci√≥n:

```python
sns.scatterplot(Z[0],Z[1],hue=label)
```

Como podemos observar, se aprecia claramente en la figura anterior que cada grupo est√° muy separado entre s√≠. Esto significa que es de esperarse que el valor obtenido para el √≠ndice de Silhouette sea un valor cercano a 1

```python
# Probemos con 3 cl√∫ster a ver qu√© pasa ahora...

KMean= KMeans(n_clusters=3)
KMean.fit(Z)
label=KMean.predict(Z)
```

```python
#Calculamos el √çndice de Silhouette
print(f'Silhouette Score(n=3): {silhouette_score(Z, label)}')
```

```python
#Ploteamos
sns.scatterplot(Z[0],Z[1],hue=label,palette='inferno_r')
```

Apreciaciones
Como podemos observar de la figura anterior, los grupos no est√°n bien separados. La distancia entre grupos entre el grupo 1 y el grupo 2 es casi insignificante. Por eso la puntuaci√≥n de silueta para n = 3 es menor que la de n = 2.
Cuando se trata de dimensiones m√°s altas, la puntuaci√≥n de silueta es bastante √∫til para validar el funcionamiento del algoritmo de agrupamiento, ya que no podemos usar ning√∫n tipo de visualizaci√≥n para validar el agrupamiento cuando las dimensiones son mayores que 3.
Tambi√©n podemos usar la puntuaci√≥n de silueta para verificar el n√∫mero √≥ptimo de cl√∫steres. En el ejemplo anterior, podemos decir que el n√∫mero √≥ptimo de cl√∫steres es 2 ya que su puntuaci√≥n de silueta es mayor que la de 3 cl√∫steres.

# Ejemplo 3

```python
#Cargamos las librerias
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Carga del data set 
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#Creacion de K-Means
km = KMeans(n_clusters=3, random_state=42)
km.fit_predict(X)
```

```python
#Calculo de Silhouette
score = silhouette_score(X, km.labels_, metric='euclidean')
print('Silhouetter Score: %.3f' % score)
```

En esta secci√≥n del notebook, usaremos YellowBrick una biblioteca de visualizaci√≥n de aprendizaje autom√°tico para dibujar las gr√°ficas de silueta y realizar un an√°lisis comparativo para identificar el n√∫mero √≥ptimo de cl√∫sters.
Yellowbrick ampl√≠a la API de Scikit-Learn para facilitar la selecci√≥n del modelo y el ajuste de hiperpar√°metros. En este ejemplo en particular, generaremos un gr√°fico de silueta para cl√∫steres de K-medias con n_cluster = 2, 3, 4 y 5.

```python
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8)) #Definimos el tama√±o del gr√°fico
for i in [2, 3, 4, 5]:
    
    #Creando diferentes instancias de kmeans para distinta cantidad de cluster
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)

    #Visualizamos el √≠ndice Silhouette con yellowbrick
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X)
```

¬øQu√© podemos interpretar de la visualizaci√≥n anterior?
Sin dudas, a medida que agregamos m√°s cl√∫sters el √≠ndice de Silhouette empeora, por ejemplo con 3 cl√∫ster posee un valor inferior a 0.6, caso contrario con dos cl√∫ster el valor obtenido supera dicho umbral.

```python
#Kmeans con k=2
km = KMeans(n_clusters=2, random_state=42)
km.fit_predict(X)
```

```python
#Calculo de Silhouette
score = silhouette_score(X, km.labels_, metric='euclidean')
print('Silhouetter Score: %.3f' % score)
```

Insight
Efectivamente con k=2 el valor de Silhouette mejora relativamente, tal cual como pod√≠amos observar en la imagen anterior.

# Ejemplo 4

```python
#Importamos las librerias!
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from warnings import filterwarnings
filterwarnings('ignore')

# Importamos la data
df = pd.read_csv('faithful.csv')
df.head()
```

```python
# Ploteamos la data!
plt.figure(figsize=(6, 6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel('Tiempo de erupci√≥n en minutos')
plt.ylabel('Tiempo de espera para la pr√≥xima erupci√≥n.')
plt.title('Visualizaci√≥n de datos')
```

```python
# Estandarizamos los datos!
X_std = StandardScaler().fit_transform(df)
```

Recordemos...

El m√©todo Elbow o tambi√©n conocido como el m√©todo del codo, nos da una idea de cu√°l ser√≠a un buen n√∫mero de k para nuestro cl√∫stering. Para dicha elecci√≥n evaluamos cual es el valor de K donde la curva empieza a aplanarse y forma la imagen de un codo.

```python
# Gr√°fiquemos
sse = [] #Suma de la distancia al cuadrado
list_k = list(range(1, 10))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(X_std)
    sse.append(km.inertia_)

# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Numero de clusters *k*')
plt.ylabel('Suma de la distancia al cuadrado')
```

```python
#Aplicacion de Kmeans
km = KMeans(n_clusters=2, random_state=42)
km.fit(X_std)
```

```python
#Calculo del √çndice de Silhouette
score = silhouette_score(X_std, km.labels_, metric='euclidean')
print('Silhouetter Score: %.3f' % score)
```

```python
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8)) #Definimos el tama√±o del gr√°fico
for i in [2, 3, 4,5]:
    
    #Creando diferentes instancias de kmeans para distinta cantidad de cluster
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)

    #Visualizamos el √≠ndice Silhouette con yellowbrick
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X_std)
```

Insight
k = 2 es el valor m√°s √≥ptimo a elegir, dado que a mayor n√∫mero de k el √≠ndice de Silhouette disminuye.
